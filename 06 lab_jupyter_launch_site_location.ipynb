Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Function to determine the marker color based on the 'class'
def assign_marker_color(launch_class):
    # Class 1 (Success) is green
    if launch_class == 1:
        return 'green'
    # Class 0 (Failure) is red
    else:
        return 'red'

# Apply the function to the 'class' column and store the result in the new 'marker_color' column
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
site_map.add_child(marker_cluster)

# Iterate through each launch record in spacex_df
for index, record in spacex_df.iterrows():
    # 1. Get the coordinate and launch site name
    coordinate = [record['Lat'], record['Long']]
    launch_site = record['Launch Site']
    color = record['marker_color'] # 'green' or 'red'
    
    # Create the popup label for the marker
    popup_text = f"Launch Site: {launch_site}<br>Outcome: {'Success' if record['class'] == 1 else 'Failure'}"
    
    # 2. Create a Marker object
    # We use folium.Icon to get a nice, color-coded marker
    marker = folium.Marker(
        coordinate,
        # Create a popup showing the launch details
        popup=popup_text,
        # Customize the Marker's icon property using the marker_color column
        icon=folium.Icon(
            color='white',            # Marker background color
            icon_color=color,         # Icon symbol color (green or red)
            icon='rocket',            # Use a rocket icon from Font-Awesome
            prefix='fa'               # Font-Awesome prefix
        )
    )
    
    # 3. Add the Marker to the cluster
    marker_cluster.add_child(marker)

# Display the map (assuming you trust the notebook!)
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closest coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
```

```python

# --- Define Coordinates ---
launch_site_lat = 28.563197
launch_site_lon = -80.576820
road_lat = 28.56321
road_lon = -80.57077
road_coordinate = [road_lat, road_lon]

# --- 1. Calculate the distance to the road ---
distance_road = calculate_distance(
    launch_site_lat,
    launch_site_lon,
    road_lat,
    road_lon
)

print(f"Distance to closest road: {distance_road:.2f} KM")

# --- 2. Create a Marker for the Road Point ---
# Use a car icon to symbolize the road/highway
road_marker = folium.Marker(
    road_coordinate,
    icon=folium.Icon(color='black', icon='car', prefix='fa')
)
site_map.add_child(road_marker)

# --- 3. Create a Distance Marker Label ---
distance_marker_road = folium.Marker(
    # Placing the distance marker slightly away from the road point for clarity
    [road_lat, road_lon + 0.0005],
    icon=DivIcon(
        icon_size=(150, 20),
        icon_anchor=(0, 0),
        # Display the distance in the required format
        html='<div style="font-size: 12; color:#000000;"><b>Road: %s</b></div>' % "{:10.2f} KM".format(distance_road),
    )
)
site_map.add_child(distance_marker_road)

# --- 4. Draw a PolyLine connecting the launch site and the road point ---
coordinates_road = [
    [launch_site_lat, launch_site_lon],
    [road_lat, road_lon]
]
lines_road = folium.PolyLine(locations=coordinates_road, weight=2, color='black', opacity=0.8)
site_map.add_child(lines_road)

site_map # Display the final map with all markers and distance lines
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
# Coordinates for the launch site (CCAFS SLC-40)
launch_site_lat = 28.563197
launch_site_lon = -80.576820
launch_site_coordinate = [launch_site_lat, launch_site_lon]

# --- Helper Function to Create Marker, Line, and Distance Label ---
def add_distance_visualization(site_coord, target_coord, icon_name, color, label):
    """Calculates distance, adds a marker, a PolyLine, and a distance label to the map."""
    
    # 1. Calculate Distance
    distance = calculate_distance(
        site_coord[0], site_coord[1],
        target_coord[0], target_coord[1]
    )
    
    print(f"Distance to {label}: {distance:.2f} KM")
    
    # 2. Create Target Marker
    folium.Marker(
        target_coord,
        icon=folium.Icon(color='white', icon_color=color, icon=icon_name, prefix='fa'),
        tooltip=label
    ).add_to(site_map)
    
    # 3. Create Distance Label Marker (using DivIcon for text)
    folium.Marker(
        # Position label near the target point
        [target_coord[0] - 0.002, target_coord[1]], 
        icon=DivIcon(
            icon_size=(150, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12; color:{color};"><b>{label}: {distance:.2f} KM</b></div>',
        )
    ).add_to(site_map)
    
    # 4. Draw PolyLine
    coordinates_line = [site_coord, target_coord]
    lines = folium.PolyLine(
        locations=coordinates_line,
        weight=2,
        color=color,
        opacity=0.6
    )
    site_map.add_child(lines)

# --- Define Infrastructure Points ---
# City (Titusville)
city_coord = [28.5872, -80.8066]

# Railway
railway_coord = [28.5721, -80.5852]

# Highway (Road)
highway_coord = [28.56321, -80.57077]

# --- Run Visualization for Each Infrastructure ---
add_distance_visualization(
    launch_site_coordinate, city_coord, 'building', 'purple', 'City'
)

add_distance_visualization(
    launch_site_coordinate, railway_coord, 'train', 'orange', 'Railway'
)

add_distance_visualization(
    launch_site_coordinate, highway_coord, 'road', 'black', 'Highway'
)

site_map # Display the final map
```
